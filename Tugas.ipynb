```python
#4
def persegi_panjang (panjang, lebar) :
    luas = panjang * lebar 
    keliling = 2 * (panjang + lebar)
    return luas, keliling

p = 7
l = 6

luas, keliling = persegi_panjang (p, l)

print("panjang : ", p)
print("lebar : ", l)
print("Luas persegi panjang :", luas)
print("Keliling Persegi panjang : ", keliling)
```
